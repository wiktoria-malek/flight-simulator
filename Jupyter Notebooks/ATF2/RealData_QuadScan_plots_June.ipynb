```python
import numpy as np
import matplotlib
import pickle, os, sys
from pathlib import Path
project_root_path = Path.cwd().resolve()
while not (project_root_path / "Backend").exists() and project_root_path.parent != project_root_path:
    project_root_path = project_root_path.parent
sys.path.insert(0, str(project_root_path))
os.chdir(project_root_path)
from Backend.State import State
```

```python
def load_quad_scan_folder(folder):
    folder = Path(folder)
    files = sorted(folder.glob("screen_*_step_*_shot_*.pkl"))
    rows = []

    for path in files:
        parts = path.stem.split("_")
        screen_i = int(parts[1])
        step_i = int(parts[3])
        shot_i = int(parts[5])

        state = State(filename=str(path))
        screen = state.get_screens()
        quad = state.get_quadrupoles()

        rows.append({
            "file": str(path),
            "screen_i": screen_i,
            "screen": screen["names"][0],
            "step": step_i,
            "shot": shot_i,
            "K1": float(np.ravel(quad["bdes"])[0]),
            "sigx": float(np.ravel(screen["sigx"])[0]) / 1000.0,
            "sigy": float(np.ravel(screen["sigy"])[0]) / 1000.0,
        })

    print(f"Loaded {len(rows)} states")
    return rows

def plot_scan(rows):
    screens = sorted(set(r["screen"] for r in rows))
    for screen in screens:
        rr = sorted([r for r in rows if r["screen"] == screen], key=lambda r: r["K1"])
        K1 = np.array([r["K1"] for r in rr])
        sigx = np.array([r["sigx"]**2 for r in rr])
        sigy = np.array([r["sigy"]**2 for r in rr])
        fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(7, 6))
        ax1.plot(K1, sigx, "o-")
        ax2.plot(K1, sigy, "o-")
        ax1.set_title(screen)
        ax1.set_ylabel("sigx^2 [mm]")
        ax2.set_ylabel("sigy^2 [mm]")
        ax2.set_xlabel("K1 [1/m²]")
        ax1.grid(True, alpha=0.3)
        ax2.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
```
